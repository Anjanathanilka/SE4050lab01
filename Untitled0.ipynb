```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-5, 5,100)
y = np.linspace(-5, 5,100)
X, Y = np.meshgrid(x, y)
plt.plot(x, y)
plt.show()
```

```python
Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_title("3D Plot of Z = X² + Y²")
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_zlabel("Z axis")

plt.show()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
```

```python
# Display first 5 observations
df.head()
```

```python
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
stats_df = df[features]
```

```python
# Pearson Correlation
pearson_corr = stats_df.corr(method='pearson')

# Spearman Correlation
spearman_corr = stats_df.corr(method='spearman')
```

```python
plt.figure(figsize=(14, 6))

# Pearson Heatmap
plt.subplot(1, 2, 1)
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Pearson Correlation")

# Spearman Heatmap
plt.subplot(1, 2, 2)
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Spearman Correlation")

plt.tight_layout()
plt.show()
```
